Como variáveis como Nível de Membresia, Método de Pagamento Preferido, Dispositivo Usado e Aceitação de Oferta influenciam o número de viagens?



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_excel('/content/perfis_usuarios_uber_10000_linhas.xlsx')

# Análise exploratória básica
print(df.describe())
print(df.info())

# Agrupamento por Nível de Membresia e visualização do Número de Viagens
sns.boxplot(data=df, x='Nivel_de_Membresia', y='Numero_de_Viagens', palette="viridis")
plt.title("Distribuição do Número de Viagens por Nível de Membresia")
plt.show()

# Tabela resumo para Nível de Membresia e Média de Aceitação de Primeira e Segunda Ofertas
membership_offer_acceptance = df.groupby('Nivel_de_Membresia')[['Media_Aceitacao_Primeira_Oferta', 'Media_Aceitacao_Segunda_Oferta']].mean()
print(membership_offer_acceptance)

# Agrupamento por Método de Pagamento e visualização do Número de Viagens
sns.barplot(data=df, x='Metodo_de_Pagamento_Preferido', y='Numero_de_Viagens', palette="magma")
plt.title("Número de Viagens por Método de Pagamento Preferido")
plt.show()
```

Existe relação entre o gasto médio por viagem e variáveis como Dispositivo, Ano de Registro e Nível de Membresia?

```python
# Análise do Gasto Médio por Viagem vs. Dispositivo Usado
sns.boxplot(data=df, x='Dispositivo_Usado', y='Gasto_Medio_por_Viagem', palette="rocket")
plt.title("Gasto Médio por Viagem por Dispositivo Usado")
plt.show()

# Gráfico de dispersão para ver a relação entre Idade e Gasto Médio por Viagem
sns.scatterplot(data=df, x='Idade', y='Gasto_Medio_por_Viagem', hue='Nivel_de_Membresia', palette="coolwarm")
plt.title("Idade vs Gasto Médio por Viagem, Segmentado por Nível de Membresia")
plt.show()

# Agrupamento por Ano de Registro e Gasto Médio por Viagem
gasto_ano = df.groupby('Ano_de_Registro')['Gasto_Medio_por_Viagem'].mean()
gasto_ano.plot(kind='line', marker='o', color='orange')
plt.title("Evolução do Gasto Médio por Viagem por Ano de Registro")
plt.ylabel("Gasto Médio por Viagem")
plt.xlabel("Ano de Registro")
plt.show()
```

```python
# Distribuição da Taxa Média de Desistência por Nível de Membresia
sns.boxplot(data=df, x='Nivel_de_Membresia', y='Taxa_Media_Desistencia', palette="cividis")
plt.title("Distribuição da Taxa Média de Desistência por Nível de Membresia")
plt.show()

# Relação entre Média de Aceitação de Primeira e Segunda Ofertas e Taxa de Desistência
sns.scatterplot(data=df, x='Media_Aceitacao_Primeira_Oferta', y='Taxa_Media_Desistencia', hue='Nivel_de_Membresia', palette="viridis")
plt.title("Aceitação de Primeira Oferta vs Taxa de Desistência")
plt.show()

# Tabela resumo da Taxa de Desistência por Nível de Membresia
desistencia_membresia = df.groupby('Nivel_de_Membresia')['Taxa_Media_Desistencia'].mean()
print(desistencia_membresia)
```

Análise do Perfil de Viagem por Faixa Etária

```python
# Criando faixas etárias
df['Faixa_Etaria'] = pd.cut(df['Idade'], bins=[18, 25, 35, 50, 65, 80], labels=['18-25', '26-35', '36-50', '51-65', '66-80'])

# Visualização do número de viagens por faixa etária
sns.boxplot(data=df, x='Faixa_Etaria', y='Numero_de_Viagens', palette="plasma")
plt.title("Distribuição do Número de Viagens por Faixa Etária")
plt.show()

# Gasto médio por viagem por faixa etária
sns.barplot(data=df, x='Faixa_Etaria', y='Gasto_Medio_por_Viagem', palette="cool")
plt.title("Gasto Médio por Viagem por Faixa Etária")
plt.show()

# Taxa de desistência por faixa etária
sns.lineplot(data=df, x='Faixa_Etaria', y='Taxa_Media_Desistencia', marker="o", color="red")
plt.title("Taxa Média de Desistência por Faixa Etária")
plt.show()
```

Análise de Tendências ao Longo do Tempo (Ano de Registro)

```python
# Evolução do número de viagens por ano de registro
sns.lineplot(data=df, x='Ano_de_Registro', y='Numero_de_Viagens', marker='o', color="green")
plt.title("Evolução do Número de Viagens por Ano de Registro")
plt.show()

# Evolução da Taxa de Desistência e do Gasto Médio por Viagem por Ano de Registro
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel("Ano de Registro")
ax1.set_ylabel("Taxa Média de Desistência", color=color)
ax1.plot(df.groupby('Ano_de_Registro')['Taxa_Media_Desistencia'].mean(), color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel("Gasto Médio por Viagem", color=color)
ax2.plot(df.groupby('Ano_de_Registro')['Gasto_Medio_por_Viagem'].mean(), color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Evolução da Taxa de Desistência e Gasto Médio por Viagem")
plt.show()
```

Análise de Preferência de Dispositivo e Desempenho




```python
# Distribuição do Número de Viagens por Dispositivo
sns.boxplot(data=df, x='Dispositivo_Usado', y='Numero_de_Viagens', palette="viridis")
plt.title("Distribuição do Número de Viagens por Dispositivo")
plt.show()

# Comparação entre Gasto Médio por Viagem e Taxa de Desistência por Dispositivo
sns.catplot(data=df, x='Dispositivo_Usado', y='Gasto_Medio_por_Viagem', kind="bar", palette="magma")
plt.title("Gasto Médio por Viagem por Dispositivo Usado")
plt.show()

sns.catplot(data=df, x='Dispositivo_Usado', y='Taxa_Media_Desistencia', kind="bar", palette="coolwarm")
plt.title("Taxa Média de Desistência por Dispositivo Usado")
plt.show()
```
